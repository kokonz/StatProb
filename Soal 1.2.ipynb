## Soal 1

Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

## Jawab :

### a). Range (Rentang)

```python
data1 = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data1}")

def calculate_range(data1):
    gol_tertinggi = max(data1)
    print(f"Nilai tertinggi: {gol_tertinggi}")

    gol_terendah = min(data1)
    print(f"Nilai terendah: {gol_terendah}")

    data_range = max(data1) - min(data1)
    
    print(f"\n")    
    print(f"Range: {gol_tertinggi} - {gol_terendah} = {data_range}")  
    
    return data_range

data_range = calculate_range(data1)
```

### Histogram

```python
import matplotlib.pyplot as plt

plt.hist(data1, bins=30, edgecolor='black', alpha=0.5, color='orange')

y = 1.5 
x1 = 4/70*1  
x2 = 66/70*1  

plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(data1)}')

plt.axvline(x=min(data1), color='blue', linestyle='solid', linewidth=1, label=f'Min Data: {min(data1):.2f}')
plt.axvline(x=max(data1), color='blue', linestyle='solid', linewidth=1, label=f'Max Data: {max(data1):.2f}')

plt.xlabel('Banyak Gol')
plt.ylabel('Frekuensi')
plt.title('Range')

plt.legend()

plt.show()
```

### b). Variance (Variansi)

```python
data1 = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data1}")

def calculate_mean(data1):
    return sum(data1) / len(data1)

def calculate_variance(data1):
    mean = calculate_mean(data1)
    print(f"Rata-rata: {mean}")  
    
    print(f"Kuadrat selisih setiap gol data dengan rata-rata:")    
    squared_diff = []
    for x in data1:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")    
    
    print(f"Total kuadrat selisih: {sum(squared_diff)}")    
    print(f"Jumlah total gol: {len(data1)}")    
    
    variance = sum(squared_diff) / len(data1)
    
    print(f"\n")    
    print(f"Variansi: {sum(squared_diff)} / {len(data1)} = {variance}")    
    
    return variance

variance = calculate_variance(data1)
```

### c). Standard Deviation (Standar Deviasi)

```python
data1 = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data1}")

def calculate_standard_deviation(data1):
    variance = calculate_variance(data1)
    standard_deviation = variance**0.5
    print(f"\n")   
    print(f"Standard Deviation: {variance}**0.5 = {standard_deviation}")
    
    return standard_deviation

standard_deviation = calculate_standard_deviation(data1)
mean = calculate_mean(data1)
```

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(data1, bins=30, edgecolor='black', alpha=0.5, color='brown')

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='solid', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='solid', linewidth=1, label=f'{mean+standard_deviation:.2f}')

plt.xlabel('Total Gol')
plt.ylabel('Frekuensi')
plt.title(f'Std Deviation = {standard_deviation:.2f}')

plt.legend()

plt.show()
```

### d). Data Baru

```python
data_baru1 = [12, 13, 15, 14, 13]
print(f"Data: {data_baru1}")

standard_deviation = calculate_standard_deviation(data_baru1)
mean_baru = calculate_mean(data_baru1)
```

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(data_baru1, bins=30, edgecolor='black', alpha=0.5, color='pink')

plt.axvline(x=mean_baru, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_baru}')
plt.axvline(x=mean_baru - standard_deviation, color='blue', linestyle='solid', linewidth=1, label=f'{mean_baru-standard_deviation:.2f}')
plt.axvline(x=mean_baru + standard_deviation, color='blue', linestyle='solid', linewidth=1, label=f'{mean_baru+standard_deviation:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation:.2f}')

plt.legend()

plt.show()
```

### e). Kesimpulan

Standar Deviasi yang besar menunjukkan tingkat variasi data yang tinggi, memberikan informasi yang lebih bervariasi. Sebaliknya, Standar Deviasi yang kecil mencerminkan tingkat konsistensi atau prediktabilitas yang lebih tinggi dalam dataset.
